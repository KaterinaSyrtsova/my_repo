1. Функция `arab2rome()`. На вход подаётся вектор `a` из натуральных чисел от 1 до 10. Необходимо вернуть вектор той же длины, элементы которого содержат в себе римские цифры.

e.g.: **Input**: `c(1, 1, 2, 3, 5, 8)`. **Output**: `c("I", "I", "II", "III", "V", "VIII")`

```r
arab2rome <- function(a) {
    roman <- c("I","II","III","IV","V","VI","VII","VIII","IX","X", NULL)
    a[a>10] <- 11
    return(roman[a])
}

arab2rome(c(1,1,2,3,5,8))
arab2rome(c(1:10))
```

2. Функция `palindrome()`. На вход поступает строка `str`, которую надо проверить, является ли она палиндромом. В случае положительного или отрицательного ответа выведете соответствующее сообщение на экран.

e.g.1: **Input**: `'"Not New York", - Roy went on'`. **Output**: `"This string is a palindrome!"`

e.g.2: **Input**: `"Hello, world!"`. **Output**: `"This string is not a palindrome!"`

```r
library(stringi)
palindrome <- function(str) {
    answer <- c("This string is not a palindrome!","This string is a palindrome!")
    str <- tolower(gsub('[^[:alnum:] ]', '', gsub(" ", "", str)))
    return(answer[identical(str,stri_reverse(str))+1])
}
palindrome("\"Not New York\", - Roy went on")
palindrome("Hello, world!")
```

3. Функция `spiral()`. Дано натуральное число `n`, целое неотрицательное число `k` и логическое значение `cw`. Необходимо построить квадратичную матрицу-спираль из элементов от 1 до $ n^2 $, где параметр ``cw`` отвечает за направление спирали: `cw == TRUE` – направление по часовой стрелке, `cw == FALSE` (по умолчанию) – направление против часовой стрелки. Параметр `k` отвечает за начальное положение спирали (e.g.: `n = 4`):
  * $ k = 0, 4, 8, \dots $; `cw == TRUE`: $$ \begin{pmatrix} 1 & 2 & 3 & 4 \\ 12 & 13 & 14 & 5 \\ 11 & 16 & 15 & 6 \\ 10 & 9 & 8 & 7 \end{pmatrix} $$
  * $ k = 0, 4, 8, \dots $; `cw == FALSE`: $$ \begin{pmatrix} 1 & 12 & 11 & 10 \\ 2 & 13 & 16 & 9 \\ 3 & 14 & 15 & 8 \\ 4 & 5 & 6 & 7 \end{pmatrix} $$
  * $ k = 1, 5, 9, \dots $; `cw == TRUE`: $$ \begin{pmatrix} 10 & 11 & 12 & 1 \\ 9 & 16 & 13 & 2 \\ 8 & 15 & 14 & 3 \\ 7 & 6 & 5 & 4 \end{pmatrix} $$
  * $ k = 1, 5, 9, \dots $; `cw == FALSE`: $$ \begin{pmatrix} 4 & 3 & 2 & 1 \\ 5 & 14 & 13 & 12 \\ 6 & 15 & 16 & 11 \\ 7 & 8 & 9 & 10 \end{pmatrix} $$
  * $ k = 2, 6, 10, \dots $; `cw == TRUE`: $$ \begin{pmatrix} 7 & 8 & 9 & 10 \\ 6 & 15 & 16 & 11 \\ 5 & 14 & 13 & 12 \\ 4 & 3 & 2 & 1 \end{pmatrix} $$
  * $ k = 2, 6, 10, \dots $; `cw == FALSE`: $$ \begin{pmatrix} 7 & 6 & 5 & 4 \\ 8 & 15 & 14 & 3 \\ 9 & 16 & 13 & 2 \\ 10 & 11 & 12 & 1 \end{pmatrix} $$
  * $ k = 3, 7, 11, \dots $; `cw == TRUE`: $$ \begin{pmatrix} 4 & 5 & 6 & 7 \\ 3 & 14 & 15 & 8 \\ 2 & 13 & 16 & 9 \\ 1 & 12 & 11 & 10 \end{pmatrix} $$
  * $ k = 3, 7, 11, \dots $; `cw == FALSE`: $$ \begin{pmatrix} 10 & 9 & 8 & 7 \\ 11 & 16 & 15 & 6 \\ 12 & 13 & 14 & 5 \\ 1 & 2 & 3 & 4 \end{pmatrix} $$

```r
right <- function(a,i,j,n,count,cw=FALSE,fg){
  a[j,(i+1):(i+n)] <- (count+1):(count+n)
  count <- count+n
  i <- i+n
  fg <- fg+1
  if (n-(fg%%2) != 0){
    if (cw){
      down(a,i,j,n-(fg%%2),count,cw, fg%%2)
    } else{
      up(a,i,j,n-(fg%%2),count,cw, fg%%2)
    }
  } else{
    print(a)
  }
}

left <- function(a,i,j,n,count,cw=FALSE,fg){
  a[j,(i-1):(i-n)] <- (count+1):(count+n)
  count <- count+n
  i <- i-n
  fg <- fg+1
  if (n-(fg%%2) != 0){
    if (cw){
      up(a,i,j,n-(fg%%2),count,cw, fg%%2)
    } else{
      down(a,i,j,n-(fg%%2),count,cw, fg%%2)
    }
  } else{
    print(a)
  }
}

down <- function(a,i,j,n,count,cw=FALSE,fg){
  a[(j+1):(j+n),i] <- (count+1):(count+n)
  count <- count+n
  j <- j+n
  fg <- fg+1
  if (n-(fg%%2) != 0){
    if (cw){
      left(a,i,j,n-(fg%%2),count,cw, fg%%2)
    } else{
      right(a,i,j,n-(fg%%2),count,cw, fg%%2)
    }
  } else{
    print(a)
  }
}

up <- function(a,i,j,n,count,cw=FALSE,fg){
  a[(j-1):(j-n),i] <- (count+1):(count+n)
  count <- count+n
  j <- j-n
  fg <- fg+1
  if (n-(fg%%2) != 0){
    if (cw){
      right(a,i,j,n-(fg%%2),count,cw, fg%%2)
    } else{
      left(a,i,j,n-(fg%%2),count,cw, fg%%2)
    }
  } else{
    print(a)
  }
}

spiral <- function(n, k, cw = FALSE) {
    m <- matrix(0,n,n)
    k <- k%%4
    k
    if (k==0){
      if (cw){
        right(m,0,1,n,0,cw,0)
      } else {
        down(m,1,0,n,0,cw,0)
      }
    }
    if (k==1){
      if (cw){
        down(m,n,0,n,0,cw,0)
      } else {
        left(m,n+1,1,n,0,cw,0)
      }
    }
    if (k==2){
      if (cw){
        left(m,n+1,n,n,0,cw,0)
      } else {
        up(m,n,n+1,n,0,cw,0)
      }
    }
    if (k==3){
      if (cw){
        up(m,1,n+1,n,0,cw,0)
      } else {
        right(m,0,n,n,0,cw,0)
      }
    }
}
```

```r
print("По часовой")
spiral(4,0,TRUE)
spiral(4,1,TRUE)
spiral(4,2,TRUE)
spiral(4,3,TRUE)
print("Против часовой")
spiral(5,4,FALSE)
spiral(5,5,FALSE)
spiral(5,6,FALSE)
spiral(5,7,FALSE)
```
