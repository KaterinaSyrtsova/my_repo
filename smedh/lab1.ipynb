## Основные теоретические положения

**Генеральная совокупность** - совокупность всех объектов (единиц), относительно которых предполагается делать выводы при изучении конкретной задачи.

**Выборочная совокупность (выборка)** - часть генеральной совокупности элементов, которая охватывается экспериментом.

**Репрезентативная выборка** - выборка, в которой все основные признаки генеральной совокупности, из которой извлечена данная выборка, представлены приблизительно в той же пропорции или с той же частотой, с которой данный признак выступает в этой генеральной совокупности.

**Главный принцип формирования репрезентативной выборки** – случайность её формирования. Выборка может быть повторной и бесповторной.

**Способы формирования выборки без разбиения генеральной совокупности на части:**
1. Простой случайный бесповторный отбор;
2. Простой случайный повторный отбор

**Способы формирования выборки с разбиением генеральной совокупности на части:**
1. Типический;
2. Механический;
3. Серийный

**Статистический ряд** - последовательность элементов выборки, расположенных в порядке их получения (наблюдения).

**Ранжированный ряд** - последовательность элементов выборки, расположенных в порядке возрастания их значений. Номер элемента ранжированного ряда в последовательности называется рангом.

**Вариационный ряд** - получается из ранжированного ряда в результате объединения одинаковых элементов. Элементы вариационного ряда называются вариантами.

**Построение интервального ряда**. Если количество вариант достаточно велико, то необходимо разделять вариационный ряд на интервалы. В этом случае ряд называется интервальным. Количество интервалов $k$ (в зависимости от объема выборки $N$) может быть вычислено по формуле Стерджесса: $k = 1 + 3.322 \lg{(N)}$
Полученное значение округляется до целого, рекомендуется выбирать нечетное число интервалов.
Ширина интервала $h$ вычисляется по формуле: $h = \frac{x_{max} - x_{min}}{k}$
Границы интервалов определяются по формуле: $[x_{min} + (i - 1)*h; x_{min} + i*h), i = 1, 2, ..., k - 1$
Последний интервал определяется по формуле: $[x_{min} + (k - 1)*h; x_{min} + k*h]$

**Эмпирической функцией распределения** называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоты события $X < x$. График представляет собой лестничный график, длина каждой ступеньки которого равна длине соответствующего интервала, а высота - отношению накопленной частоты до середины этого интервала к объему выборки.

### Выбор двумерной генеральной совокупности

В качестве генеральной совокупности был выбран датасет forestfires.csv (https://archive.ics.uci.edu/ml/datasets/Forest+Fires), содержащий информацию о лесных пожарах, для двумерной совокупности выбраны параметры DMC - индекс по системе FWI и temp - температура в градусах Цельсия.

### Формирование выборки

Из генеральной совокупности в соответствии с вариантом была сформирована выборка случайным образом с помощью функции sample. Полученая выборка из 105 пар (файл forestfires_105.csv) представлена ниже:

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

data = pd.read_csv('forestfires.csv', usecols=['DMC', 'temp'], sep=',')
data = data.sample(105)
data.to_csv('forestfires_105.csv', sep=',')
data
```

### Построение ранжированного, вариационного и интервального рядов

#### Первый признак (DMC)
##### Ранжированный ряд:

```python
data_DMC = data.loc[:,['DMC']]
ranked_DMC = data_DMC.sort_values(by='DMC')
ranked_DMC
```

Ранжированный ряд является упорядоченным по неубыванию, минимальный элемент равен 2.4, максимальный - 290.0.

##### Вариационный ряд:

```python
variation_DMC = pd.DataFrame(ranked_DMC.groupby('DMC').size())
variation_DMC = variation_DMC.reset_index()
variation_DMC.columns = ['DMC', 'frequency']
variation_DMC
```

Для получения вариационного ряда были удалены повторяющиеся элементы ранжированного ряда. Значения DMC достаточно разнообразны, частота появления большинства значений равна 1, также встречается значение с частотой равной 4.

##### Интервальный ряд:

Для формирования интервального ряда, вариационный ряд был разделен на интервалы, количество которых вычислено по формуле Стерджесса $ k = 1 + 3.222 \lg{(N)} $

Для каждого интервала найдена абсолютная частота попаданий соответствующего признака в интервал.

```python
k = 1 + math.floor(3.322 * math.log10(ranked_DMC.size))
h_DMC = (ranked_DMC['DMC'].max() - ranked_DMC['DMC'].min()) / k
points = np.concatenate((np.arange(ranked_DMC['DMC'].min(), ranked_DMC['DMC'].max(), h_DMC), [ranked_DMC['DMC'].max()]))
interval_DMC = ranked_DMC.groupby(pd.cut(ranked_DMC['DMC'], points)).size().reset_index(name='frequency')
interval_DMC
```

Частота попадания значений в отрезки интервального ряда разнообразна, большинство значений сосредоточены в первой половине отрезков, так в третьем и четвертом отрезке находится наибольшее количество элементов выборки - 27, а в шестом наименьшее - 3.

#### Второй признак (temp)

##### Ранжированный ряд:

```python
data_temp = data.loc[:,['temp']]
ranked_temp = data_temp.sort_values(by='temp')
ranked_temp
```

Ранжированный ряд является упорядоченным по неубыванию, минимальный элемент равен 4.6, максимальный - 32.4. Интервал значений гораздо меньше чем для первого признака.

##### Вариационный ряд:

```python
variation_temp = pd.DataFrame(ranked_temp.groupby('temp').size())
variation_temp = variation_temp.reset_index()
variation_temp.columns = ['temp', 'frequency']
variation_temp
```

Значения temp также разнообразны.

##### Интервальный ряд:

```python
h_temp = (ranked_temp['temp'].max() - ranked_temp['temp'].min()) / k
points = np.concatenate((np.arange(ranked_temp['temp'].min(), ranked_temp['temp'].max(), h_temp), [ranked_temp['temp'].max()]))
interval_temp = ranked_temp.groupby(pd.cut(ranked_temp['temp'], points)).size().reset_index(name='frequency')
interval_temp
```

По интервальному ряду второго признака можно сделать вывод, что температура имеет нормальное распределение.

### Построение полигона, гистограммы и эмпирической функции для интервального ряда абсолютных частот

Полигон относительных частот интервального ряда представляет собой ломанную, соединяющую точки $(x_i, n_i)$, где $x_i$ - середины соответствующих интервалов, а $n_i$ - абсолютная частота попадания в соответствующий интервал.

#### Полигон для интервального ряда асболютных частот DMC

```python
polygon_DMC = pd.DataFrame()
polygon_DMC['DMC'] = interval_DMC['DMC'].apply(lambda x: x.mid).astype('float')
polygon_DMC['n'] = interval_DMC['frequency']
polygon_DMC.plot(x='DMC', y='n', ylabel='n', title='Polygon absolute frequency DMC')
```

#### Гистограмма для интервального ряда асболютных частот DMC

Для построения гистограммы были построены столбцы с шириной основания, равной ширине интервала, и высотой, равной отношению частоты попадания вариант в соответствующий интервал к ширине интервала. Таким образом, для гистограммы абсолютных частот площадь столбцов равна объему выборки, а для гистограммы относительных частот – единице.

```python
hist_DMC, bins_DMS = np.histogram(ranked_DMC['DMC'], bins=k)
hist_DMC = hist_DMC.astype('float') / h_DMC
plt.title('Histogram absolute frequency DMC')
plt.xlabel('DMC')
plt.ylabel('n/h')
plt.hist(bins_DMS[:-1], bins_DMS, weights=hist_DMC)
```

Гистограмма и полигон отражают характер распределения величины признака по интервалам.

#### Эмпирическая функция для интервального ряда абсолютных частот DMC

```python
cumsum_DMC = np.concatenate(([0], np.cumsum(polygon_DMC['n'])))
plt.title('Empirical function absolute frequency DMC')
plt.xlabel('DMC')
plt.ylabel('F*(DMC)')
plt.step(x=bins_DMS, y=cumsum_DMC, where='mid')
```

#### Полигон для интервального ряда асболютных частот temp

```python
polygon_temp = pd.DataFrame()
polygon_temp['temp'] = interval_temp['temp'].apply(lambda x: x.mid).astype('float')
polygon_temp['n'] = interval_temp['frequency']
polygon_temp.plot(x='temp', y='n', ylabel='n', title='Polygon absolute frequency temp')
```

#### Гистограмма для интервального ряда асболютных частот temp

```python
hist_temp, bins_temp = np.histogram(ranked_temp['temp'], bins=k)
hist_temp = hist_temp.astype('float') / h_temp
plt.title('Histogram absolute frequency temp')
plt.xlabel('temp')
plt.ylabel('n/h')
plt.hist(bins_temp[:-1], bins_temp, weights=hist_temp)
```

#### Эмпирическая функция для интервального ряда абсолютных частот temp

```python
cumsum_temp = np.concatenate(([0], np.cumsum(polygon_temp['n'])))
plt.title('Empirical function absolute frequency temp')
plt.xlabel('temp')
plt.ylabel('F*(temp)')
plt.step(x=bins_temp, y=cumsum_temp, where='mid')
```

### Построение полигона, гистограммы и эмпирической функции для интервального ряда относительных частот частот
#### Полигон для интервального ряда относительных частот DMC

```python
sum_DMC = interval_DMC['frequency'].sum()
polygon_DMC['w'] = interval_DMC['frequency'] / sum_DMC
polygon_DMC.plot(x='DMC', y='w', ylabel='w', title='Polygon relative frequency DMC')
```

#### Гистограмма для интервального ряда относительных частот DMC

```python
plt.title('Histogram relative frequency DMC')
plt.xlabel('DMC')
plt.ylabel('w/h')
plt.hist(bins_DMS[:-1], bins_DMS, weights=hist_DMC / sum_DMC)
```

#### Эмпирическая функция для интервального ряда относительных частот DMC

```python
plt.title('Empirical function relative frequency DMC')
plt.xlabel('DMC')
plt.ylabel('F*(DMC)')
plt.step(x=bins_DMS, y=cumsum_DMC / sum_DMC, where='mid')
```

#### Полигон для интервального ряда относительных частот temp

```python
sum_temp = interval_temp['frequency'].sum()
polygon_temp['w'] = interval_temp['frequency'] / sum_temp
polygon_temp.plot(x='temp', y='w', ylabel='w', title='Polygon relative frequency temp')
```

#### Гистограмма для интервального ряда относительных частот temp

```python
plt.title('Histogram relative frequency temp')
plt.xlabel('temp')
plt.ylabel('w/h')
plt.hist(bins_temp[:-1], bins_temp, weights=hist_temp / sum_temp)
```

#### Эмпирическая функция для интервального ряда относительных частот temp

```python
plt.title('Empirical function relative frequency temp')
plt.xlabel('temp')
plt.ylabel('F*(temp)')
plt.step(x=bins_temp, y=cumsum_temp / sum_temp, where='mid')
```

При сравнении графиков (полигон, гистограмма, эмпирическая функция), построенных для интервальных рядов абсолютных и относительных частот, можно сделать вывод, что форма графиков не изменяется, изменяются значения по оси $y$ в соответствии с объемом выборки.
По характеру графиков можно сделать вывод о распределении признаков, первый признак имеет левостороннюю ассиметрию, а второй признак имеет распределение близкое к нормальному.
